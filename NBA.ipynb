```python
top_players=df1.sort_values(by='pts',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='pts',y='player_name',data=top_players,palette='viridis',hue='player_name')
plt.title('Top 10 Players')
plt.xlabel('pts per game')
plt.ylabel('player name')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='ast', y='pts', hue='team_abbreviation', palette='tab10')
plt.title('Points vs Assists (by Team)')
plt.xlabel('Assists')
plt.ylabel('Points')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['pts'], bins=20, kde=True, color='purple')
plt.title('Distribution of Points Per Game')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```

```python
top_rebounders = df1.sort_values(by='reb', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='reb', y='player_name', data=top_rebounders, palette='Greens')
plt.title('Top 10 Players by Rebounds')
plt.xlabel('Rebounds Per Game')
plt.ylabel('Player Name')
plt.show()
```

```python
players = df1[df1['player_name'].isin(['LeBron James', 'Stephen Curry'])]
sns.lineplot(data=players, x='season', y='pts', hue='player_name')
plt.show()
```
